Concatenation

```python
X = torch.arange(12, dtype = torch.float32).reshape((3, 4))
Y = torch.tensor([[2.0, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])
X, Y
```

```python
X.shape, Y.shape
```

```python
torch.cat((X, Y), dim = 0)
```

```python
torch.cat((X, Y), dim = 1)
```
